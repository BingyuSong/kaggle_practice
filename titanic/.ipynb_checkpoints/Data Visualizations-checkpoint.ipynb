```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
df=pd.read_csv("train1.csv")
```

```python
sex=df['Sex']
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(sex)
sex1=le.transform(sex)
Embarked=df['Embarked']
Embarked=Embarked.tolist()
le.fit(Embarked)
Embarked=le.transform(Embarked)
```

```python
del df['Sex']
df.insert(4,'Sex',sex1)
del df['Embarked']
df.insert(11,'Embarked',Embarked)
```

```python
del df['Name']    
```

```python
del df['Ticket']
```

```python
del df['Cabin']
```

```python
%matplotlib inline 
for i in df:
    pt=df.plot(kind="scatter", x="Survived", y=i)
    
```

```python
for i in df:
    sns.jointplot(x="Survived", y=i, data=df, size=5)
```

```python
import numpy as np
f, ax = plt.subplots(figsize=(12, 9))
# corr = d.corr()
df=pd.read_csv('data.csv')
corr = df.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, ax=ax,annot=True)
```
